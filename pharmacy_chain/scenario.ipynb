# Pharmacy Chain Discrete Event Simulation
## Scenario
### Intro
The Pharmacy Chain has `N` pharmacies scattered in 2 areas:
1. City Center
2. City Periphery
3. Rural

All locations share the same Warehouse, which in this simulation has an "infinite" amount of medication.

### Medication Types
All the medication is separated into 2 main categories: *OTC* (Over the Counter) and *Prescription* medication.

OTC Medication used in this simulation:
- Ibuprofen (painkiller, antipyretic)
- Paracetamol (painkiller, antipyretic)
- Vitamin C (flu and general wellness)
- Vitamin D (wellness)
- Zinc (wellness)
- Calcium carbonate (reflux)
- Benzydamine (sore throat)
- Amylmetacresol (sore throat)
- Loperamide (antidiarrheal)
- Oxymetazoline (decongestant)

Prescription medication:
- Amoxicillin (antibiotic)
- Enalapril (cardio-vascular)
- Metformin (diabetes)
- Escitalopram (psychiatric)
- Alprazolam (psychiatric)
- Tramadol (prescription painkiller) 

### Customer Personas
Customers are generated with different "personas". This affects the probabilities of the medication they will want to buy.

- **Quick-Fixer**: only buys 1-2 OTCs for a certain cause. Easily accepts alternatives. Low-Medium patience.
- **Wellness Customer**: buys vitamins. High patience.
- **Acute Flu Patient**: buys OTCs and/or prescription medication for colds and flus.
- **Senior**: chroinic patient that gets the same medication.
- **Depressive**: chronic patient. Buys psychiatric medication and sometimes vitamins.
- **Post-Op Patient**: rare. Buys antibiotics and serios painkillers.
- **Stockpiler**: buys many OTCs at once, for the whole family. 
- **Problem Patient**: somewhat rare. Arrives with expired prescriptions or no prescription. 

### Stock management.
At the end of every work day, the stock is checked and if it is below a certain `threshold`, a restock from the warehouse is ordered for the next morning. The truck with the medication takes some time to arrive, based on the area the pharmacy is in.
